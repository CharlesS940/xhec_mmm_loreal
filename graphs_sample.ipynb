```python
# Data for the categories
categories = ['Base', 'TV', 'Display', 'Search', 'Social', 'Other']
cy2022_values = [62721, 15000, 5000, 6000, 3000, 2000]
cy2023_values = [72364, 18000, 6000, 7000, 3500, 2500]

# Colors in shades of pink
colors = ['#FFC0CB', '#FFB6C1', '#FF69B4', '#FF1493', '#C71585', '#DB7093']

# Create a figure
fig, axes = plt.subplots(1, 2, figsize=(16, 8), subplot_kw=dict(aspect="equal"))

# Function to create a donut chart
def create_donut_chart(ax, values, title, total_revenue, incremental_sales):
    wedges, texts = ax.pie(values, wedgeprops=dict(width=0.3), startangle=90, colors=colors)
    ax.set_title(title)

    # Adding total revenue and incremental sales
    ax.text(0, 0, f'{total_revenue} M [REVENUE]\nIncremental Sales: {incremental_sales}%',
            horizontalalignment='center', verticalalignment='center', fontsize=12, color='black')

    # Adding labels outside the donut chart to avoid overlap
    for i, wedge in enumerate(wedges):
        angle = (wedge.theta2 - wedge.theta1) / 2.0 + wedge.theta1
        y = 1.1 * np.sin(np.deg2rad(angle))  # Move labels slightly outside the donut
        x = 1.1 * np.cos(np.deg2rad(angle))
        percentage = values[i] / sum(values) * 100
        ax.text(x, y, f'{categories[i]}\n{percentage:.0f}%', 
                horizontalalignment='center', verticalalignment='center', fontsize=9, color='black')

# Create donut charts for CY2022 and CY2023
create_donut_chart(axes[0], cy2022_values, 'CY 2022', 95, 33)
create_donut_chart(axes[1], cy2023_values, 'CY 2023', 114, 37)

# Add legends to the side for better readability
axes[0].legend(categories, loc='upper left', fontsize=10)
axes[1].legend(categories, loc='upper left', fontsize=10)

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
