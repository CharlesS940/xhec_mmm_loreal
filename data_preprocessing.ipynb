```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ap_variables = pd.read_csv("data/a&p_variables.csv")
commercial_variables = pd.read_csv("data/commercial_variables.csv")
kpi_variables = pd.read_csv("data/kpi_to_model.csv")
```

```python
# Convert 'Starting Week' to datetime
ap_variables["Starting week"] = pd.to_datetime(ap_variables["Starting week"])
commercial_variables["Starting Week"] = pd.to_datetime(
    commercial_variables["Starting Week"]
)
kpi_variables["Starting Week"] = pd.to_datetime(kpi_variables["Starting Week"])

# rename column in ap_variables
ap_variables = ap_variables.rename(columns={"Starting week": "Starting Week"})
```

```python
# Merge the datasets on 'Starting Week'
merged_data = pd.merge(
    ap_variables,
    commercial_variables,
    how="left",
    on="Starting Week",
)
merged_data = pd.merge(merged_data, kpi_variables, how="left", on="Starting Week")
```
