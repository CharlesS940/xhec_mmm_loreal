```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ap_variables = pd.read_csv("data/a&p_variables.csv")
commercial_variables = pd.read_csv("data/commercial_variables.csv")
kpi_variables = pd.read_csv("data/kpi_to_model.csv")
```

```python
# Convert 'Starting Week' to datetime
ap_variables["Starting week"] = pd.to_datetime(ap_variables["Starting week"])
commercial_variables["Starting Week"] = pd.to_datetime(
    commercial_variables["Starting Week"]
)
kpi_variables["Starting Week"] = pd.to_datetime(kpi_variables["Starting Week"])

# rename column in ap_variables
ap_variables = ap_variables.rename(columns={"Starting week": "Starting Week"})
```

```python
# Merge the datasets on 'Starting Week'
merged_data = pd.merge(
    ap_variables,
    commercial_variables,
    how="left",
    on="Starting Week",
)
merged_data = pd.merge(merged_data, kpi_variables, how="left", on="Starting Week")
```

```python
# Calculate total sales
merged_data['total_sales'] =merged_data['UK L\'Oreal Paris Haircare Total Offline Sellout Value (in pound)'] +merged_data['UK L\'Oreal Paris Haircare Total Online Sellout Value (in pound)']

# Create a new DataFrame to focus on sales by channel
sales_data =merged_data[['Year_x', 'UK L\'Oreal Paris Haircare Total Offline Sellout Value (in pound)', 
                 'UK L\'Oreal Paris Haircare Total Online Sellout Value (in pound)', 'total_sales']].copy()

# Plot
plt.figure(figsize=(10,6))
sales_data.set_index('Year_x')[['UK L\'Oreal Paris Haircare Total Offline Sellout Value (in pound)', 
                                'UK L\'Oreal Paris Haircare Total Online Sellout Value (in pound)']].plot(kind='bar', stacked=True)

plt.title('Sales by Channel (Offline vs Online)')
plt.ylabel('Sales in Pounds')
plt.xlabel('Year')
plt.xticks(rotation=45)
plt.legend(title='Sales Channel', labels=['Offline Sales', 'Online Sales'], loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Create the stacked area plot with light pink and dark pink colors
plt.figure(figsize=(12, 6))
plt.stackplot(kpi_variables['Starting Week'], 
              kpi_variables['Offline Sales'], 
              kpi_variables['Online Sales'], 
              labels=['Offline Sales', 'Online Sales'],
              colors=['#FFB6C1', '#FF69B4'])  # Light pink and dark pink

# Add labels, title, and legend
plt.title('Weekly Sales by Channel (Offline vs Online)', fontsize=16)
plt.xlabel('Week', fontsize=12)
plt.ylabel('Sales (in pounds)', fontsize=12)
plt.legend(loc='upper left', fontsize=12)

# Format x-axis to show dates neatly
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.grid(True, linestyle='--', alpha=0.6)  # Add gridlines for clarity

# Show the plot
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
